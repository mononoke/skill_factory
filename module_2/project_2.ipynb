```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('files/stud_math.csv')
```

```python
df.head(10)
```

```python
## Буду записывать важные пункты в массив
result = list()


## Массив для признаков, у которых не видно влияния на оценку
can_be_removed = list()
```

```python
## Функция для вывода основной информации по колонке
def print_column_info(column):
    count_not_empty = df[column].count()
    count_empty = len(df) - count_not_empty
    p_empty = count_empty *100 / len(df)

    
    print('-- Уникальные значения', df[column].unique())
    print('-- Количество уникальных значений', df[column].nunique())
    print('-- Количество непустых значений', count_not_empty)
    print('-- Количество пустых значений', count_empty)
    print('-- Процент пустых значений', p_empty, '%')
    print('-- Значения в процентном соотношении\n', df[column].value_counts(normalize = True))  
    print('-- Показатели оценок\n', df.groupby([column])['score'].describe())


## Функция для вывода моды оценок по признаку
def print_column_score_mode(column):
    c = df[column].value_counts().reset_index()
    print(c)
    for i in c.index:
        name = c.loc[i]['index']
        print('-- Мода', name, ': ', df[df[column] == name]['score'].mode())


## Функция для bar-графика распределения признака
def bar_plot(column, title, xlabel):
    return sns.countplot(
        x = column,
        data = df
    ).set(
        title = title,
        xlabel = xlabel,
        ylabel = 'Количество'
    )


## Функция для отрисовки bar по распределению оценок
def bar_score_plot(columns, title, aggfunc = 'count'):
    return df.pivot_table(
        columns = columns,
        index = 'score',
        aggfunc = aggfunc,
        values = 'score2'
    ).plot(
        kind = 'bar',
        title = title,
        ylabel = 'Количество',
        xlabel = 'Оценка',
        figsize = (20, 5),
        grid = True,
        width = 1
    )


## Функция для отрисовки box по распределению оценок
def box_plot(x, title, xlabel):
    return sns.boxplot(
        x = x,
        y = 'score',
        data = df
    ).set(
        title = title,
        ylabel = 'Оценка',
        xlabel = xlabel
    )
```

Датафрейм:

- по факту настоящих числовых значений только три (score, age и absences). Все остальное - кодирование признака числом

### score - баллы по госэкзамену по математике

```python
df['score2'] = df.score


df.score.hist().set(
    title = 'Распределени баллов',
    xlabel = 'Баллы',
    ylabel = 'Количество'
)
df.score.describe()
```

```python
## Учитывая, что оценка по госэкзамену - основной показатель нашего исследования,
## и пропущенных значений мало, то я их просто удалю
df = df[df['score'].notna()]
```

```python
## соотношение признаков для нулевых баллов
for column in df.columns:
    if (column != 'score' and column != 'score2'):
        df[df['score'] == 0.0].pivot_table(
            columns = column,
            index = 'score',
            aggfunc = 'count',
            values = 'score2'
        ).plot(
            kind = 'bar',
            title = column,
            ylabel = 'Количество',
            xlabel = 'Оценка',
            figsize = (20, 5),
            grid = True,
            width = 1
        )
```

```python
## Посмотрим на оценки в процентах
df.score.value_counts(normalize=True)
```

- Для оценки по госэкзамену по математике используется стобальная система
- Из 395 значений 6 пропущено - я их отфильтровываю и не использую в дальнейшем
- Видно, что 9.5% процентов получили 0 баллов. Следующий бал - 20 баллов и его получили только 0.2%.  
0 баллов может говорить о том, что это ошибка в датасете, но также может свидетельствовать о том, что есть система недопуска к экзамену или система с минимальными требованиями выполнения.  
Видно, что все, у кого стоит 0 баллов не пропускали ни одного занятия. Может 0 баллов - это автомат?  
На данном этапе с нулевыми оценками нет ясности.
- Высший балл получили 0.2%, что соответствует социальным нормам во всем мире.
- Оценок выше 100 или ниже 0 нет. Все ОК
- 50% оценок лежит между 40 и 70 баллами, что также является социальной нормой

```python
absences_q1 = df['absences'].quantile(0.25)
absences_q3 = df['absences'].quantile(0.75)
IQR = absences_q3 - absences_q1


## Выбросы
min_v = absences_q1 - IQR * 1.5
max_v = absences_q3 + IQR * 1.5


print('Q1: ', absences_q1)
print('Q3: ', absences_q3)
print('IQR', IQR)
print('min_v', min_v)
print('max_v', max_v)
```
